```python
A = np.array((1.0,1.0))
B = np.array((2.0,0.5))
C = np.array((1.0,2.5))
D = np.array((3.0,3.5))
E = np.array((5.5,3.5))
F = np.array((5.5,2.5))
```

```python
Datapoints = {'A':A, 'B':B,'C':C,'D':D,'E':E,'F':F}
```

```python
Datapoints['A'] - Datapoints['B']
```

```python
def l1_distance(dp):
    L1 = np.array([])
    for x in Datapoints:
        dist = round(np.linalg.norm(Datapoints[dp] - Datapoints[x]), 1 )
        L1 = np.append(L1, dist)
    return print(L1)
```

```python

for x in Datapoints:
    l1_distance(x)
```

```python

def l2_distance(dp):
    L2 = np.array([])
    for x in Datapoints:
        dist = round(np.linalg.norm(Datapoints[dp] - Datapoints[x]), 2 )
        L2 = np.append(L2, dist)
    return print(L2)
```

```python
#L2 = np.array([])
for y in Datapoints:
    l2_distance(y)
    
```
